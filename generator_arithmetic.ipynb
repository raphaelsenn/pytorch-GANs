## Generator Arithmetic

```python
import torch
import matplotlib.pyplot as plt

from DCGAN.dcgan_celeb_faces import Generator
```

```python
seed = 10000
torch.manual_seed(seed)
```

```python
weights = torch.load('./generator_celebA_weights.pth')
generator = Generator()
generator.load_state_dict(weights)
```

```python
z1 = torch.distributions.uniform.Uniform(-1, 1).sample((1, 100))
z1
```

```python
IMAGE_SHAPE = (3, 64, 64)
```

```python
img = generator(z1)
img = img.view(IMAGE_SHAPE).permute(1, 2, 0).detach().numpy()
plt.imshow(((img + 1)/2).clip(0, 1))
```

```python
z2 = torch.distributions.uniform.Uniform(-1, 1).sample((1, 100))
```

```python
img = generator(z2)
img = img.view(IMAGE_SHAPE).permute(1, 2, 0).detach().numpy()
plt.imshow(((img + 1)/2).clip(0, 1))
```

```python
z3 = torch.distributions.uniform.Uniform(-1, 1).sample((1, 100))
img = generator(z3)
img = img.view(IMAGE_SHAPE).permute(1, 2, 0).detach().numpy()
plt.imshow(((img + 1)/2).clip(0, 1))
```

```python
z4 = z1 - z2 + z3
img = generator(z4)
img = img.view(IMAGE_SHAPE).permute(1, 2, 0).detach().numpy()
plt.imshow(((img + 1)/2).clip(0, 1))
```
